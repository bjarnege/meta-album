```python
import pandas as pd
import numpy as np
```

```python
training_loss_acc = pd.read_csv("experiment_results_train.csv")
valid_loss_acc = pd.read_csv("experiment_results_validation.csv")[:-500]
```

```python
training_loss_acc["running_loss_training"] = training_loss_acc["Loss"].rolling(100).mean()
training_loss_acc["running_acc_training"] = training_loss_acc["Accuracy"].rolling(100).mean()
```

```python
training_loss_acc["epoch"] = np.ceil((training_loss_acc.index+0.000001) / 1000).astype(int)
valid_loss_acc["epoch"]  = np.ceil((valid_loss_acc.index+0.000001) / 500).astype(int)
valid_per_epoch = valid_loss_acc.groupby("epoch").mean()
train_per_epoch = training_loss_acc.groupby("epoch").last()[["running_loss_training",	"running_acc_training"]]
```

```python
epochs_df = valid_per_epoch.join(train_per_epoch).rename(columns={"Accuracy": "acc_validation",
                                                                  "Loss": "loss_validation"})
```

```python
for type in ["loss", "acc"] + list(epochs_df.columns): 
    epochs_df[[col for col in epochs_df.columns if type in col]].plot()
```
